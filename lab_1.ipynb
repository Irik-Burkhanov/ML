```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt


df = pd.read_csv('student-mat.csv')
df.info()
```

```python
# Количество обучающихся по возрасту
df.groupby('age')['age'].count()
```

```python
# Средний возраст среди учащихся, употребляющих алкоголь в будни
df.groupby('Dalc')['age'].mean()
```

```python
# Средний возраст среди учащихся, употребляющих алкоголь в выходные
df.groupby('Walc')['age'].mean()
```

```python
# Среднее качество семейных отношений среди учащихся, употребляющих алкоголь в будни (от 1-плохое до 5-отличное)
df.groupby('Dalc')['famrel'].mean()
```

```python
# Среднее качество семейных отношений среди учащихся, употребляющих алкоголь в выходные (от 1-плохое до 5-отличное)
df.groupby('Walc')['famrel'].mean()
```

```python
# Среднее количество пропусков среди студентов, употребляющих алкоголь в будни
df.groupby('Dalc')['absences'].mean()
```

```python
# Среднее количество пропусков среди студентов, употребляющих алкоголь в выходные
df.groupby('Walc')['absences'].mean()
```

```python
# Средняя годовая оценка среди студентов, употребляющих алкоголь в будни (максимум 20)
df.groupby('Dalc')['G3'].mean()
```

```python
# Средняя годовая оценка среди студентов, употребляющих алкоголь в выходные (максимум 20)
df.groupby('Walc')['G3'].mean()
```

```python
# Общее количество студентов, употребляющих алкоголь в будни
df.groupby('Dalc').count()
```

```python
# Общее количество студентов, употребляющих алкоголь в выходные
df.groupby('Walc').count()
```

```python
# Гистограмма потребления алкоголя в зависимости от свободного времени в будни
df.groupby('Dalc')[['freetime']].mean().plot(kind='bar')
```

```python
# Гистограмма потребления алкоголя в зависимоти от свободного времени в выходные
df.groupby('Walc')[['freetime']].mean().plot(kind='bar')
```
